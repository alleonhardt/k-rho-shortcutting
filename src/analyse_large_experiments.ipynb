```python
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
def add_column_postifix(df, postifix):
    dc = {}
    for x in df.columns:
        dc[x] = x+postifix
    return df.rename(columns=dc)
```

```python
result = pd.read_csv("../paper/src/data/hyperbolic_large.csv")
rhos = [lambda x: x-1, lambda x: int(x/2), lambda x: int(x/3), lambda x: int(x/4), lambda x: int(x/10), lambda x: int(math.sqrt(x))]
result["regime"] = [0 for x in range(len(result))]
result = result.drop(["build_hash","generator"],axis=1)

for k in [3,4,5]:
    k_cond = result["k"]==k
    for (id,rh) in enumerate(rhos):
        is_rho = result["n"].apply(rh)==result["rho"]
        result.loc[k_cond&is_rho,"regime"] = len(rhos)-id
result["approx"] = result["size_bl_dp"]/result["size_bl_dp_pert_min_pc"]
result["approx_pert"] = result["size_bl_dp_pert_min"]/result["size_bl_dp_pert_min_pc"]
result["approx_pc"] = result["size_bl_dp_pc_min"]/result["size_bl_dp_pert_min_pc"]
result["approx_ms"] = result["size_bl_dp_s"]/result["size_bl_dp_pert_min_pc"]
result["approx_greedy"] = result["size_bl_dp_g"]/result["size_bl_dp_pert_min_pc"]

for k in [3,4,5]:
    onlyk = result[result["k"]==k].groupby("regime")
    res = pd.concat([onlyk.mean(["approx","approx_ms","approx_pert","approx_pc","approx_greedy"]), add_column_postifix(2*onlyk.sem(),"_error")],axis=1)
    res.to_csv(f"results_comparison_hyperbolic_{k}.csv")
```

```python
result = pd.read_csv("../paper/src/data/real_world.csv")
col = result.columns.tolist()
col.remove("size_bl_dp")
col.remove("size_bl_dp_pert_min_pc")
col.remove("n")
result = result.drop(col,axis=1)
result["approx"] = result["size_bl_dp"]/result["size_bl_dp_pert_min_pc"]
result = result.dropna()

for n in [2500,5000,7500,10000]:
    partial = result[(result["n"]<=n) & (result["n"] > (n-2500))]
    partial.to_csv(f"../paper/src/data/real_world_proc_{n}.csv",index=False)
    fr=partial[["approx","n"]].mean().to_frame().transpose()
    fr.to_csv(f"../paper/src/data/real_world_proc_{n}_mean.csv", index=False)
```
